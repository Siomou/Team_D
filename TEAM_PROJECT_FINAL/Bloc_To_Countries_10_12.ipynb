```python
#Δώσαμε το path για το αρχείο
file_path = 'C:\\Users\\jason\\Documents\\GitHub\\cmcprojectpersonal\\eurovision_song_contest_1957_2023.xlsx'
```

```python
#Διαβάσαμε το αρχείο
df = pd.read_excel(file_path)
```

```python
#Αντικαταστήσαμε τα ονόματα κάποιον χωρών
df['From country'] = df['From country'].replace({'F.Y.R. Macedonia': 'North Macedonia', 'Macedonia': 'North Macedonia'})
df['To country'] = df['To country'].replace({'F.Y.R. Macedonia': 'North Macedonia', 'Macedonia': 'North Macedonia'})

df['From country'] = df['From country'].replace({'Yugoslavia': 'F.U.Y/S.M.', 'Serbia & Montenegro': 'F.U.Y/S.M.'})
df['To country'] = df['To country'].replace({'Yugoslavia': 'F.U.Y/S.M.', 'Serbia & Montenegro': 'F.U.Y/S.M.'})
```

```python
#Αλλάξαμε τα ονόματα των columns
df = df.rename(columns={'From country': 'from_country',
                                         'To country':'to_country',
                                         'Points      ':'points'})
```

```python
#Kάναμε νέο column στο df και ορίσαμε τα ονόματα των χωρών που θα αντικατασταθούν με τα ονόματα των μπλοκ
df['from_bloc'] = ''
```

```python
#Ορίσαμε τα ονόματα των χωρών που θα αντικατασταθούν με τα ονόματα των μπλοκ
country_to_bloc_mapping = {
    "Sweden" : "Nordic Bloc",
    "Norway" : "Nordic Bloc",
    "Iceland" : "Nordic Bloc",
    "Finland" : "Nordic Bloc",
    "Denmark": "Nordic Bloc",
    
    "Andorra" : 'Iberian Bloc',
    "Spain" : 'Iberian Bloc',
    "Portugal" : 'Iberian Bloc',
    
    "Armenia" : 'East Bloc',
    "Azerbaijan" : 'East Bloc',
    "Belarus" : 'East Bloc',
    "Czech Republic" : 'East Bloc',
    "Estonia" : 'East Bloc',
    "Georgia" : 'East Bloc',
    "Latvia" : 'East Bloc',
    "Lithuania" : 'East Bloc',
    "Poland" : 'East Bloc',
    "Russia" : 'East Bloc',
    "Slovakia" : 'East Bloc',
    "Ukraine" : 'East Bloc',
    
    "Austria" : 'Western Bloc',
    "Belgium" : 'Western Bloc',
    "France" : 'Western Bloc',
    "Germany" : 'Western Bloc',
    "Ireland" : 'Western Bloc',
    "Luxembourg" : 'Western Bloc',
    "Netherlands" : 'Western Bloc',
    "Switzerland" : 'Western Bloc',
    "United Kingdom" : 'Western Bloc',
    
    "Albania" : 'Balkan Bloc', 
    "Bosnia & Herzegovina" : 'Balkan Bloc',
    "Bulgaria" : 'Balkan Bloc',
    "Croatia" : 'Balkan Bloc',
    "Cyprus": 'Balkan Bloc',
    "North Macedonia" : 'Balkan Bloc',
    "Greece" : 'Balkan Bloc',
    "Hungary" : 'Balkan Bloc',
    "Moldova" : 'Balkan Bloc',
    "Montenegro" : 'Balkan Bloc',
    "Sebia" : 'Balkan Bloc',
    "F.U.Y/S.M." : 'Balkan Bloc',
    "Slovenia" : 'Balkan Bloc',
    "Turkey" : 'Balkan Bloc',
    "Romania": 'Balkan Bloc'
}
```

```python
#Κάναμε την αντικατάσταση και τα βάλαμε στο νέο column
for index, row in df.iterrows():
    country = row['from_country']
    bloc = country_to_bloc_mapping.get(country, 'Other Countries')
    df.at[index, 'from_bloc'] = bloc
```

```python
#Αλλάξαμε τη σειρά των columns
desired_columns_order = ["Year", "(semi-) final", "Edition", "Jury or Televoting", "from_country", "from_bloc", "to_country", "points", "Duplicate"]
df = df[desired_columns_order]
```

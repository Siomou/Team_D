### Τα Δεδομένα για τις Βαθμολογίες της Eurovision

Αναζητήσαμε και βρήκαμε δεδομένα που περιέχουν τις βαθμολογίες από όλες τις διοργανώσεις της Eurovision, ενώ δημιουργήσαμε και έναν κώδικο scraping, εργαλείο που διδαχθήκαμε κατά τη διάρκεια του εξαμήνου και επιθυμούσαμε να δοκιμάσουμε.

Τα δεδομένα που χρησιμοποιήσαμε αντλήθηκαν από την ιστοσελίδα data.world. Πρόκειται για ένα παγκόσμιο αποθετήριο δεδομένων ανοιχτής πρόσβασης. Ένας από τους χρήστες του αποθετηρίου, βρήκε τα δεδομένα από το eurovision.tv, την επίσημη ιστοσελίδα της Eurovision και αφού τα επεξεργάστηκε (σ.σ. «καθάρισε») και επαλήθευσε την ορθότητα και την ακρίβειά τους, τα ανέβασε στον προσωπικό του λογαριασμό στο αποθετήριο data.world.

Πρώτο Dataset  (eurovision_song_contest_1957_2023 (με φίλτρα))

Ο πίνακας έχει συγκεντρωμένες όλες τις βαθμολογίες που αποδόθηκαν από χώρα σε χώρα από την έναρξη του διαγωνισμού έως και σήμερα και αποτέλεσε την αρχική πηγή πληροφοριών μας. Αυτό αποτέλεσε το αρχικό μας γενικό dataset, το οποίο αργότερα επεξεργαστήκαμε. Πάνω σε αυτό βασίστηκε η αρχική μας έρευνα και ο εντοπισμός κάποιων μοτίβων. Το εν λόγω dataset είναι αρκετά εκτενές καθώς περιλαμβάνει 61.819 γραμμές και 8 στήλες.
Στήλη Α [Year]: Το έτος που διεξήχθη ο εκάστοτε διαγωνισμός. Περιλαμβάνει αριθμητικές τιμές που αντιστοιχούν σε χρονολογίες.
Στήλη B [(semi-)final]: Το στάδιο του διαγωνισμού κατά το οποίο απονέμεται η κάθε βαθμολογία. Περιλαμβάνει μία εκ τεσσάρων επιλογών: sf = ημιτελικός, sf1 = πρώτος ημιτελικός, sf2 = δεύτερος ημιτελικός και f = τελικός.
Στήλη C [edition]: To έτος και το στάδιο του διαγωνισμού που αποδίδεται η εκάστοτε βαθμολογία. Αποτελεί επί της ουσίας έναν συνδυασμό των πρώτων 2 στηλών.
Στήλη D [Jury or Televoting]: Αναφέρεται στην προέλευση της βαθμολογίας και περιέχει 2 επιλογές. Εάν η βαθμολογία προέρχεται από την κριτική επιτροπή της χώρας, τότε εμφανίζεται η ένδειξη Jury, ενώ εάν η βαθμολογία προέρχεται από το κοινό, εμφανίζεται η ένδειξη Televoting. 
Σημείωση: Δεν υπήρχε σε κάθε διαγωνισμό η διαδικασία του Televoting.
Στήλη E [From country]: H χώρα από την οποία απονέμεται η βαθμολόγια.
Στήλη F [To country]: Η χώρα που λαμβάνει τη βαθμολογία.
Στήλη G [Points]: Οι πόντοι που απονέμονται κάθε φορά. Περιλαμβάνει αριθμητικές τιμές από 0-12.
Στήλη Η [Duplicate]: Στη στήλη αυτή εμφανίζεται ένα Χ όταν και στις 2 στήλες με το όνομα της χώρας που δίνει και λαμβάνει τους βαθμούς, εμφανίζεται η ίδια χώρα. Αυτό γίνεται καθώς στον διαγωνισμό δεν υπάρχει η δυνατότητα κάποια χώρα να ψηφίσει τον εαυτό της. Παρόλα αυτά, πιθανότατα επειδή ο δημιουργός του dataset δεν μπορούσε ή δεν ήθελε κατά την επεξεργασία του, να σβήσει αυτά τα δεδομένα, έκανε αυτή την επιλογή.

Δεύτερο Dataset (BLOC_TO_COUNTRIES_10_12.csv)

Ο παραπάνω πίνακας αποτελεί προϊόν επεξεργασίας του πρώτου dataset. Σε αυτόν παρατηρούνται 2 σημαντικές μετατροπές σε σχέση με τον αρχικό. Πρώτον, έχει προστεθεί η στήλη from_bloc και δεύτερον έχουν συμπεριληφθεί μόνο οι βαθμολογίες 10 και 12.
Στήλη Α [Year]: Το έτος που διεξήχθη ο εκάστοτε διαγωνισμός. Περιλαμβάνει αριθμητικές τιμές που αντιστοιχούν σε χρονολογίες.
Στήλη B [(semi-)final]: Το στάδιο του διαγωνισμού κατά το οποίο απονέμεται η κάθε βαθμολογία. Περιλαμβάνει μία εκ τεσσάρων επιλογών: sf = ημιτελικός, sf1 = πρώτος ημιτελικός, sf2 = δεύτερος ημιτελικός και f = τελικός.
Στήλη C [Edition]: To έτος και το στάδιο του διαγωνισμού που αποδίδεται η εκάστοτε βαθμολογία. Αποτελεί επί της ουσίας έναν συνδυασμό των πρώτων 2 στηλών.
Στήλη D [Jury or Televoting]: Αναφέρεται στην προέλευση της βαθμολογίας και περιέχει 2 επιλογές. Εάν η βαθμολογία προέρχεται από την κριτική επιτροπή της χώρας, τότε εμφανίζεται η ένδειξη Jury, ενώ εάν η βαθμολογία προέρχεται από το κοινό, εμφανίζεται η ένδειξη Televoting. 
Σημείωση: Δεν υπήρχε σε κάθε διαγωνισμό η διαδικασία του Televoting.
Στήλη E [from_country]: H χώρα που από την οποία απονέμεται τη βαθμολόγια.
Στήλη F [(from_bloc]: Το μπλοκ χωρών στο οποίο εντάσσεται η χώρα που απονέμει τους βαθμούς.
Στήλη G [to_country]: Η χώρα που λαμβάνει τη βαθμολογία.
Στήλη Η [points]: Οι πόντοι που απονέμονται κάθε φορά. Περιλαμβάνει αριθμητικές τιμές από 10 ή 12.
Σημείωση 1: Για τη στήλη (I) που έχει τίτλο [Duplicate] ισχύει ακριβώς ό,τι και στον προηγούμενο πίνακα. Δεν εντοπίζεται κανένα Χ όμως, κάτι λογικό εφόσον δε γίνεται μία χώρα να ψηφίσει τη συμμετοχή της.
Σημείωση 2: Η στήλη (J) με τίτλο [to_bloc] είναι επίσης κενή. Σε αυτήν θα αναφέρονταν το μπλοκ της χώρας που λαμβάνει τους βαθμούς. Έως αυτό το στάδιο της μελέτης ωστόσο, δεν είχαμε πραγματοποιήσει τη συγκεκριμένη ομαδοποίηση ακόμα, η οποία έγινε σε νέο dataset.

Αυτό το dataset το χρησιμοποιήσαμε για να βρούμε τους υψηλούς ψήφους (10-12) που έδωσε κάθε μπλοκ σε κάθε χώρα. Οι ψήφοι εμφανίζονται σε απόλυτο αριθμό ενώ σημειώνεται πως απουσιάζουν οι χρονιές 1956-1971 και το 1974, αφού δε δόθηκαν οι συγκεκριμένες βαθμολογίες.


Τρίτο Dataset (DATASET PER_BLOCtoBLOC.csv)

Ο παραπάνω πίνακας είναι αποτέλεσμα εκτενούς και αναλυτικής επεξεργασίας του πρώτου dataset. Αποτελείται από 3 στήλες και είναι το τελικό προϊόν στο οποίο στηρίχτηκαν κατά βάση τα συμπεράσματα στα οποία καταλήξαμε ως ομάδα. Περιλαμβάνει το ποσοστό πόντων που απονέμεται από κάθε μπλοκ στο ίδιο και τα υπόλοιπα μπλοκ και καταδεικνύει τη σχέση τους.
Πιο αναλυτικά:
Στήλη Α [from_bloc]: Το μπλοκ από το οποίο απονέμεται το ποσοστό των βαθμών.
Στήλη B [to_bloc]: Το μπλοκ το οποίο λαμβάνει το ποσοστό των βαθμών.
Στήλη C [proportion]: To ποσοστό των βαθμών που απονέμεται κάθε φορά. Περιλαμβάνει αριθμητικές τιμές με ακρίβεια 1 δεκαδικού ψηφίου.

Αυτό το dataset το χρησιμοποιήσαμε για να βρούμε τους υψηλούς ψήφους (10-12) που έδωσε κάθε μπλοκ σε κάθε μπλοκ, με τους ψήφους να εμφανίζονται με μορφή ποσοστού.

Τα παραπάνω σύνολα δεδομένων αποτελούν τη βάση για το άρθρο που θα δημοσιευθεί με τίτλο «Η Γεωπολιτική της Eurovision» και πρόκειται να δημοσιευθούν ώστε να είναι προσβάσιμα από όλους όσους επιθυμούν να τα μελετήσουν και να τα αξιοποιήσουν με κάθε τρόπο.

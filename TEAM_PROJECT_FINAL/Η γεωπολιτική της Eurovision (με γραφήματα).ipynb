**<div style='text-align: center;'> Η γεωπολιτική της Eurovision </div>**

Ο πανευρωπαϊκός διαγωνισμός τραγουδιού, γνωστός ως **Eurovision**, ξεκίνησε ως εγχείρημα απευθείας μετάδοσης προγράμματος μεταξύ ευρωπαϊκών χωρών αλλά εξελίχθηκε σε θεσμό, η φήμη του οποίου έχει περάσει τα σύνορα της Γηραιάς Ηπείρου. 

Ως θεσμός πολιτιστικής διπλωματίας προάγει τη συναδέλφωση των λαών της Ευρώπης, κατηγορείται όμως κι ότι εξυπηρετεί πολιτικά και γεωπολιτικά συμφέροντα επειδή αποτελεί και μέσο ήπιας ισχύος. Μέσω του πολιτισμού δηλαδή περνάει μηνύματα για τη woke agenda και τη διεθνή πολιτική αντί να είναι απλώς ένας διαγωνισμός τραγουδιού.



Κατά καιρούς ξεσπάνε διάφορα σκάνδαλα που αργά ή γρήγορα ξεχνιούνται, ένα όμως απασχολεί -και διχάζει- το κοινό διαχρονικά: η συστηματική ανταλλαγή βαθμών μεταξύ χωρών με γεωγραφικούς, κοινωνικοπολιτικούς ή πολιτισμικούς δεσμούς. 

Στα σχεδόν 68 χρόνια της ιστορίας του διαγωνισμού της Eurovision έχουν εντοπιστεί από ερευνητές και τους λάτρεις του θεσμού **5 μπλοκ χωρών** που θεωρείται ότι ψηφίζουν κυρίως χώρες της ομάδας τους σε σταθερή βάση, χωρίς να αποτελούν κριτήριο γι’ αυτό μόνο τα τραγούδια. Το **Σκανδιναβικό**, το **Ιβηρικό**, το **Ανατολικό**, το **Δυτικό** και το **Βαλκανικό**. </div>


**<div style='text-align: center;'> Τα μπλοκ συμμαχιών </div>**

Στο **Σκανδιναβικό** μπλοκ ανήκει η Σουηδία, η Νορβηγία, η Δανία, η Ισλανδία κι η Φινλανδία. [Από γεωγραφική άποψη](https://mirc.ntua.gr/sites/default/files/scandinavian.pdf), ως σκανδιναβικές χώρες λογίζονται η Σουηδία, η Δανία κι η Νορβηγία, αλλά κι η Φινλανδία λόγω του βορειοδυτικού τμήματός της που εκτείνεται ως τις Σκανδιναβικές Άλπεις. Η Ισλανδία συμπεριλαμβάνεται για ιστορικούς λόγους, επειδή οι Ισλανδοί κατάγονται από τους Βίκινγκ. 

Στο **Ιβηρικό** μπλοκ ανήκει η Ισπανία, η Πορτογαλία κι η Ανδόρρα λόγω γεωγραφίας και πολιτιστικών δεσμών, όχι όμως κι η Γαλλία που από τον τρόπο που ψηφίζει κλίνει προς το Δυτικό. 


Το **Ανατολικό** μπλοκ προέκυψε τόσο από γεωγραφικά όσο και από ιστορικοπολιτικά και πολιτιστικά αίτια μετά τη διάλυση της ΕΣΣΔ στις 26 Δεκεμβρίου 1991 από πρώην σοβιετικές δημοκρατίες που εντάχθηκαν στο θεσμό: την Αρμενία, το Αζερμπαϊτζάν, τη Λευκορωσία, την Εσθονία, τη Γεωργία, τη Λετονία, τη Λιθουανία, τη Ρωσία και την Ουκρανία και οι οποίες συμμάχησαν μουσικά με την Πολωνία, αλλά και την Τσεχία και τη Σλοβακία -μετά τη διάλυση της Τσεχοσλοβακίας το 1993-. Η Εσθονία αν και πρώην σοβιετική δημοκρατία και μέλος του Ανατολικού μπλοκ αποτελεί παράλληλα δορυφόρο της Φινλανδίας -λόγω γεωγραφικής εγγύτητας, γλώσσας, εθνικών ύμνων με κοινή μελωδία και άλλων πολιτιστικών στοιχείων- με την οποία ψηφίζονται συστηματικά.

Η Αυστρία, το Βέλγιο, η Γαλλία, η Γερμανία, το Λουξεμβούργο, η Ολλανδία, η Ελβετία, η Ιρλανδία και το Ηνωμένο Βασίλειο σχημάτισαν το **Δυτικό** μπλοκ λόγω ισχυρών ιστορικών και πολιτιστικών δεσμών που πηγάζουν από τη γειτνίασή τους, τις κοινές γλώσσες τους και τη [μουσική βιομηχανία](https://www.historyextra.com/period/20th-century/eurovision-when-first-history-politics-voting-scandals-winnners/).

Στο **Βαλκανικό** μπλοκ του θεσμού κατατάσσονται η Σερβία, η Βοσνία-Ερζεγοβίνη, η Βόρεια Μακεδονία, η Ελλάδα, η Κροατία, η Ρουμανία, η Αλβανία, η Βουλγαρία, το Μαυροβούνιο, η Σλοβενία, η Μολδαβία, η Τουρκία, η Κύπρος κι η Ουγγαρία.

Η Μολδαβία αν και πρώην σοβιετική δημοκρατία, από ιστορικοπολιτική αλλά και πολιτιστική άποψη -κυρίως λόγω γλώσσας- συνδέεται περισσότερο με τη Ρουμανία που ανήκει γεωγραφικά στα Βαλκάνια, οπότε κάποιοι ορισμοί την περιλαμβάνουν κι αυτήν. 

Εξαιτίας του μολδαβορουμανικού ενωτισμού εξάλλου -υπάρχουν σκέψεις να επανενωθεί με τη Ρουμανία-, ανακήρυξε την ανεξαρτησία της η Υπερδνειστερία το 1990. Έτσι αποτελεί μέλος του Βαλκανικού μπλοκ, ενεργεί ως δορυφόρος της Ρουμανίας στο θεσμό, ενώ η σχέση τους θυμίζει αυτήν της Κύπρου με την Ελλάδα. 

Η Τουρκία θεωρείται ότι ανήκει στο Βαλκανικό μπλοκ για γεωγραφικούς λόγους, για το τμήμα της που είναι γνωστό ως [«Ευρωπαϊκή Τουρκία»](http://users.auth.gr/users/1/9/050491/public_html/Greece-Europe-World/countries/turkey-html/european%20part/oria/oria.htm) και περιλαμβάνει την Ανατολική Θράκη και τη Δυτική Κωνσταντινούπολη, ενώ η Σλοβενία -αν και ανήκει στην Κεντρική Ευρώπη- τοποθετείται στα Βαλκάνια για γεωγραφικούς λόγους αφού ήταν μια από τις έξι σοσιαλιστικές δημοκρατίες που απάρτιζαν την πρώην Γιουγκοσλαβία προτού διαλυθεί την περίοδο 1991-1992.

Η Ουγγαρία γεωγραφικά μπορεί να ανήκει στην Κεντρική Ευρώπη, οι πολιτιστικοί και ιστορικοί δεσμοί της με τα Βαλκάνια όμως και οι προτιμήσεις της κατά τις ψηφοφορίες την κατέταξαν στο Βαλκανικό μπλοκ.
